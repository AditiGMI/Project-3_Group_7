```python
import requests
import json
```

```python
url = "https://archive-api.open-meteo.com/v1/archive?latitude=40.71&longitude=74&start_date=2010-01-01&end_date=2019-12-31&daily=temperature_2m_max&temperature_unit=fahrenheit&timezone=America%2FNew_York"
response = requests.get(url)
```

```python
if response.status_code == 200:
    
    data = response.json()
    
    json_data = json.dumps(data, indent=4)
    
    print(json_data)
else:
    print(f"failed")
```
